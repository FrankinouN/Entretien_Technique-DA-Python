```python
# Modifie le nom de la colonne « M » par « sexe », et la colonne « technician » par « metier », la colonne « 85711 » par « code_postal »


df.columns
```

```python
df.columns = (['1', '24', 'sexe', 'metier', 'code_postal'])
```

```python
df.columns
```

```python
#Dans la colonne sexe modifie les valeurs « F » par « femme » et « M » par « homme »

df.sexe.replace({'F': 'femme', 'M': 'homme'})
```

```python
# Supprime la colonne « 1 » et « 24 »

df.drop(['1', '24'], axis=1, inplace=True)
df
```

```python
# Crée une nouvelle colonne qui ne sera remplie que des 2 premiers chiffres de la colonne « code_postal », tu l’appelera « departement »


df.dtypes
```

```python
liste=[]
for i in df.code_postal:
    liste.append(i[:2])  
df['departement']=liste
df
```

# Fait en sorte que la colonne code_postal soit en float et la colonne département en int


```python
# Fonction trouver sur Stackoverflow avec l'aide de Sacha pour retirer des caracteres de la colonne Code postal

df['code_postal'] = df['code_postal'].str.extract('(\d+)', expand=False)
df
```

```python
df.code_postal.astype(float, copy=True, errors='ignore')
df
```

```python
# Idem pour departement

df['departement'] = df['departement'].str.extract('(\d+)', expand=False)
```

```python
df.departement.astype(float, copy=True, errors='ignore')
df
```

```python
# Affiche le pourcentage de remplissage de toutes les colonnes en ordre décroissant

def maFonction(df):
    x = df.count()
    y = x.index
    z = x*100/len(df)
    i = 0
    while i < len(x):
        print(z[i],y[i])
        i +=1
    
maFonction(df)

# sort_values
# toutes les colonnes sont à 100% .....
```
